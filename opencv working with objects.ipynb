```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("image.jpg")
```

```python
plt.imshow(img)
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
latter=cv2.CascadeClassifier("C:/Users/manda/Downloads/Telegram Desktop/haarcascade_frontalface_default.xml")
```

```python
found=latter.detectMultiScale(img_gray,minSize=(20,20))
amount_found=len(found)
```

```python
if amount_found !=0:
    for (x,y,width,height) in found:
        cv2.rectangle(img_rgb,(x,y),(x+height,y+width),(0,255,0),2)
```

```python
plt.subplot(1,1,1)
plt.imshow(img_rgb)
plt.show()
```
